```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option("display.max_colwidth", 2500)
pd.set_option("display.max_rows", 2500)
```

```python
df = pd.read_csv("top200_2024/2024_top200_clustered_v1.csv")
print(df.shape)
```

```python
"""
Overall summary
"""

cluster_summary = df.groupby(['Cluster','ClusterDescription']).agg(
    Number_of_Papers=('Title', 'count'),
    Most_Cites=('Cites', 'max'),
    Time_Range=('Year', lambda x: f"{int(x.min())} - {int(x.max())}")
).reset_index().sort_values(by=['Number_of_Papers'], ascending=[False])

display(cluster_summary)
```

```python
"""
Top N cited paper in each cluster
"""

TOP_N = 3
(
    df.sort_values(by=['Cluster','ClusterDescription','Cites'], ascending=[True, True, False])
    .groupby(['Cluster','ClusterDescription']).head(TOP_N).reset_index()
)[['Cluster','ClusterDescription','Title','Year','Authors','Source','ArticleURL','Cites']].set_index([
    'Cluster', 'ClusterDescription', 'Title'
])
```
